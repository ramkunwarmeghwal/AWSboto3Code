```python
##All EC2 Instanace accross all regions
import boto3
import csv
from pprint import pprint

boto3.setup_default_session(profile_name='staging')

ec2_cli = boto3.client('ec2')

collect_all_region = []

for region in ec2_cli.describe_regions()['Regions']:
    #print(region['RegionName'])
    collect_all_region.append(region['RegionName'])
#print(collect_all_region)

fo = open("ec2_report.csv",'w',newline='')
data_obj = csv.writer(fo)
data_obj.writerow(['Sno ','InstanceType  ','Private IP  ', 'Public IP  ', 'Launch Time'])

cnt = 1
for each_region in collect_all_region:
    ec2_rec= boto3.resource('ec2',region_name=each_region)
    for each_ins_in_reg  in  ec2_rec.instances.all():
        data_obj.writerow(([cnt , each_ins_in_reg.instance_id  , each_ins_in_reg.instance_type  ,    
                each_ins_in_reg.private_ip_address ,  each_ins_in_reg.public_ip_address  , 
                each_ins_in_reg.launch_time]))
        cnt = cnt+1
fo.close()    
print("Finished")
```

```python
import boto3
from pprint import pprint

import os

boto3.setup_default_session(profile_name='staging')

client = boto3.client('ec2')
```

```python
# create a list where the volume ids of unused volumes will be stored
volumes_to_delete = list()
volume_size = []

volume_detail = client.describe_volumes()

# process each volume in volume_detail
if volume_detail['ResponseMetadata']['HTTPStatusCode'] == 200:
    for each_volume in volume_detail['Volumes']:
      # the volumes which do not have 'Attachments' key and their state is 'available' is considered to be unused
        if len(each_volume['Attachments']) == 0 and each_volume['State'] == 'available':
            volumes_to_delete.append(each_volume['VolumeId'])
            volume_size.append(each_volume['Size'])
# these are the candidates to be deleted by maintaining waiters for them
#pprint(volumes_to_delete)
#print(volume_size)
print("Volumes are not attched list")
for i in range(len(volumes_to_delete)):
     print(volumes_to_delete[i] + ": "+ str(volume_size[i]))
```
