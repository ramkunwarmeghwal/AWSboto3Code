```python
## List all free IP and Storing in Elasticsearch
import boto3
from elasticsearch import Elasticsearch
es = Elasticsearch("192.168.255.245:9200")
boto3.setup_default_session(profile_name='staging')

client = boto3.client('ec2')
addresses_dict = client.describe_addresses()
#print(addresses_dict)
ip_dict = {} 
ip_list = []

for eip_dict in addresses_dict['Addresses']:
    if "NetworkInterfaceId" not in eip_dict:
        #print(eip_dict['PublicIp'])
        ip_list.append(eip_dict['PublicIp'])
for i in range(len(ip_list)):
    ip_dict[i+1] = ip_list[i]
print(ip_dict) 
es.index(index="ip_address", doc_type="address", id=1, body=ip_dict)
```

```python
import boto3
from pprint import pprint
from elasticsearch import Elasticsearch
es = Elasticsearch("192.168.255.245:9200")

import os

boto3.setup_default_session(profile_name='staging')

client = boto3.client('ec2')
```

```python
# create a list where the volume ids of unused volumes will be stored
volumes_to_delete = list()
volume_size = []

volume_detail = client.describe_volumes()

# process each volume in volume_detail
if volume_detail['ResponseMetadata']['HTTPStatusCode'] == 200:
    for each_volume in volume_detail['Volumes']:
      # the volumes which do not have 'Attachments' key and their state is 'available' is considered to be unused
        if len(each_volume['Attachments']) == 0 and each_volume['State'] == 'available':
            volumes_to_delete.append(each_volume['VolumeId'])
            volume_size.append(each_volume['Size'])
# these are the candidates to be deleted by maintaining waiters for them
pprint(volumes_to_delete)
print(volume_size)
# print("Volumes are not attched list")
# for i in range(len(volumes_to_delete)):
#      print(volumes_to_delete[i] + ": "+ str(volume_size[i]))

##convert above both list in a dictonary
res1 = {}
for key in volumes_to_delete:
    for value in volume_size:
        res1[key] = value
        volume_size.remove(value)
        break 

#Push this dictonary in ElasticSearch        
es.index(index="volume", doc_type="storage", id=1, body=res1)
```

```python
## List of all Instance Which are under utilization
import boto3
import sys
from datetime import datetime, timedelta


from elasticsearch import Elasticsearch
es = Elasticsearch("192.168.255.245:9200")
        
boto3.setup_default_session(profile_name='staging')

client = boto3.client('cloudwatch')
response = client.get_metric_statistics(
            Namespace='AWS/EC2',
            MetricName='CPUUtilization',
         #   Dimensions=[
          #      {
           #     'Name': 'InstanceId',
            #    'Value': 'i-1234abcd'
             #   },
            #],
            StartTime=datetime(2022, 1, 5) - timedelta(seconds=1200),
            EndTime=datetime(2022, 1, 14),
            Period=600,
            Statistics=[
                'Maximum',
            ],
            Unit='Percent'
        )


#pprint.pprint(response)
l1 = []
l2 = []
for cpu in response['Datapoints']:
    if 'Maximum' in cpu:
        if cpu['Maximum'] < 20:
          #  print(cpu['Maximum'])
            l1.append(cpu['Maximum'])
        else:
            l2.append(cpu['Maximum'])
#
#pprint.pprint(l1)
#print(l2)


#pprint.pprint(response)

#for cpu in response['Datapoints']:
 #   print(cpu)

cpu_dict = {}

for i in range(len(l1)):
    cpu_dict[i+1] = l1[i]

pprint.pprint(cpu_dict)

es.index(index="cpu", doc_type="cpu_uses", id=1, body=cpu_dict)

```
